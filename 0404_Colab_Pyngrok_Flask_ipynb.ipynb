```python
from google.colab import userdata # 導入模組，用於讀取隱密的環境變數

# 從 Colab 的秘密設定 (Secrets) 中獲取名為 'NGROK_AUTHTOKEN' 的權杖
# 這樣做可以避免將帳號密碼直接寫死在程式碼中，提升安全性
ngrok_authtoken = userdata.get('NGROK_AUTHTOKEN')

# 設定要對外開放服務的通訊埠 (Port)
port = 5051
```

```python
import os  # 導入作業系統模組，常用於存取環境變數或檔案系統
from pyngrok import ngrok  # 導入 pyngrok 套件，用於控制 Ngrok 進行通道 (tunnel) 建立
```

```python
ngrok.kill()
```

```python
# 使用先前從 Google Colab Secrets 讀取到的權杖 (token) 來驗證身分
# 這是使用 ngrok 服務所必需的步驟
ngrok.set_auth_token(ngrok_authtoken)

# 建立一個連線通道 (Tunnel)
# 將本地指定的 port (例如 5051) 映射到公網，並命名為 "linebot_tunnel" 以利識別
tunnel = ngrok.connect(port, name="linebot_tunnel")

# 印出產生的公開 URL
# 這就是您可以提供給 LINE Developers Console 的 Webhook 位址
print(tunnel.public_url)
```

```python
from flask import Flask # 導入 Flask 框架，用於建立 Web 伺服器

# 建立 Flask 應用程式實例，__name__ 代表當前的模組名稱
app = Flask(__name__)

# 定義路由 (Route)：當使用者造訪網址根目錄 '/' 時，執行下方的 home 函式
@app.route('/')
def home():
    # 回傳 HTML 格式的字串，這是瀏覽器會看到的內容
    return "<h1>B12090076 鍾翟恩連線成功</h1>"

# 確保只有在直接執行此腳本時才會啟動伺服器
if __name__ == '__main__':
    # 啟動伺服器，並將監聽埠口設定為 5000
    app.run(port=5000)
    
```
